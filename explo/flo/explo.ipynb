```python
def check_language(start, end):
    for date in pd.date_range(start, end, freq='15T').to_pydatetime():
        str_date = date.strftime("%Y%m%d%H%M%S")
        mentions = pd.read_csv(f"../../data/raw/{str_date}.mentions.csv", delimiter="\t", header=None, names=header_mentions, usecols=['MentionDocTranslationInfo']).squeeze()
        not_na_count = mentions.notna().sum()
        # print(not_na_count)
        if not_na_count > 0:
            print(f"Pas d'anglais à cette date: {date}")
        
```

```python
start = dt(2023, 1, 7)
end = dt(2023, 1, 8)
check_language(start, end)
```
